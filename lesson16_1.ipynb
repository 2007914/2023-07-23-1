```python
import numpy as np
import matplotlib.pyplot as plt
index =np.arange(4)
value=np.array([5,7,8,9,])*1000
figure =plt.figure(figsize=(8,5))
axes = figure.add_subplot(1,1,1)
axes.bar(index,value)#長條圖
axes.set_xticks(index,['gigi','mimi','gay','nigor'])
axes.grid(axis='y')
xmin,xmax,ymin,ymax=axes.axis()
print(xmin,xmax,ymin,ymax)
axes.axis([xmin,xmax,ymin,ymax])
for i in index:
    print(i,value[1],end='')
    axes.annotate(str(value[i]),[i,value[i]])
plt.show        




```
