```python
scores = np.random.randint(60,100,size=(10,5))
students = pd.DataFrame(scores,index=range(1,11),columns=["國文","英文","數學","地理","歷史"])
students['總分']= students.sum(axis=1)
students
```

```python
students.iloc[:,:5]
```

```python
students['平均']=students.iloc[:,:5].mean(axis=1)
students
```

```python
students.to_csv('學生第一學期分數.csv')
```
