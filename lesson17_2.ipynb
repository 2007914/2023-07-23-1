```python
import requests
url = 'https://opendata.cwb.gov.tw/fileapi/v1/opendataapi/F-C0032-001?Authorization=rdec-key-123-45678-011121314&format=JSON'
response = requests.get(url)
if response.status_code == 200:
    print("下載成功")
else:
    print("下載失敗")
```

```python
import json
a = response.json()
b=a['cwaopendata']['dataset']['location']
weather= []
for item in  b:
    city={}
    city['城市']= item['locationName']
    city['起始時間']=item['weatherElement'][1]['time'][0]['startTime']
    city['結束時間']=item['weatherElement'][1]['time'][0]['endTime']
    city['最高溫度']=item['weatherElement'][1]['time'][0]['parameter']['parameterName']
    city['最低溫度']=item['weatherElement'][2]['time'][0]['parameter']['parameterName']
    city['感覺'] = item['weatherElement'][3]['time'][0]['parameter']['parameterName']
    weather.append(city)
   
weather    
```

```python
import pandas as pd
df =pd.DataFrame(weather)
df.info()
```
