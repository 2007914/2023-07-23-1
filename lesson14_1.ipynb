```python
url = 'https://opendata.cwb.gov.tw/fileapi/v1/opendataapi/F-C0032-001?Authorization=rdec-key-123-45678-011121314&format=JSON'
import requests
 
response=requests.get(url )
if response.status_code == 200:
    print("下載成功")
    print(response.text)
```

```python
response.text
with open ('最新的氣象預測.json',mode='w',encoding='utf-8') as file:
    file.write(response.text)

```

```python
import json
a = response.json()
b=a['cwbopendata']['dataset']['location']
weather= []
for item in  b:
    city={}
    city['城市']= item['locationName']
    city['起始時間']=item['weatherElement'][1]['time'][0]['startTime']
    city['結束時間']=item['weatherElement'][1]['time'][0]['endTime']
    city['最高溫度']=item['weatherElement'][1]['time'][0]['parameter']['parameterName']
    city['最低溫度']=item['weatherElement'][2]['time'][0]['parameter']['parameterName']
    city['感覺'] = item['weatherElement'][3]['time'][0]['parameter']['parameterName']
    weather.append(city)
   
weather    
```

```python
import csv
with open('最新的氣象預測.csv',mode='w',encoding='utf8',newline='') as file:
    fieldnames = ['城市', '起始時間', '結束時間', '最高溫度', '最低溫度', '感覺']
    dict_writer = csv.DictWriter(file,fieldnames=fieldnames)
    dict_writer.writeheader()
    dict_writer.writerows(weather)
```
