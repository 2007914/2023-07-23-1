```python
import numpy as np
import pandas as pd

df = pd.read_csv('學生第一學期分數.csv')
```

```python
df.columns = ["學號","國文","英文","數學","地理","歷史","總分","平均"]
df
```

```python
df1 = df.rename(columns={'Unnamed': '學號'})
df1
```

```python
display(df1['總分'])
df1['總分'].rank(ascending=False,method='first')
```

```python
df1['排名'] = df1['總分'].rank(ascending=False,method= 'first')
df1
```

```python
df2 = df1.sort_values(by='排名')
```

```python
display(df2.iloc[:5])
display(df2.iloc[-5:])
```
